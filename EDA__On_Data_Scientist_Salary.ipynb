```python
def top15_barplot(df, column):
    top15_counts = df[column].value_counts().head(15)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=top15_counts.values, y=top15_counts.index, palette='viridis')
    plt.xlabel('Count')
    plt.ylabel(column)
    plt.title(f'Top 15 {column} Counts')
    plt.show()

top15_barplot(df, 'Job Title')
```
