```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(3)
```

```python
credits.head(3)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies['genres'][0] # we have to extract the value of name for recommendation
```

```python
import ast  # Abstract Syntax Trees
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
# Extract the vale of name from genres column
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['genres'][0] # Now it looks perfect
```

```python
movies['keywords'][0] # we have to extract the keywords from dictionary
```

```python
movies['keywords'] = movies['keywords'].apply(convert)   # Same we have to do with keywords
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
# we are doing this because we want to remove space between cast, crew, genres and keywords to make thing uniform and it also help machine to not to confuse
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

This part is the main step for vectorization

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

we combine all the words to make a sea of words and we can take 5000 similar words for comparison

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

CountVectorizer is the technique to featch 5000 or any amount we provied from the sea of words related to movies


```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

So it will be like row and column of movies and words, number will be assigned to them which words is mostly used

```python
from sklearn.metrics.pairwise import cosine_similarity
```

instead of euclidean distance we will use cosine_similarity because it will help us to fine the 5 nearest movie using angle between them


```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

movie which is compared to itself has a same vector space so angle is zero thats why first row first column and digonal element is 1.

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

This line gives the five vectors close to the given movie

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Thor')
```

Here we go we recomended five movie related to last watched or searched movie
