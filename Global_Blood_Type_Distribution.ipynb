```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/bloodtypes.csv')
```

```python
df.dropna(inplace = True)
```

```python
# Aggregate the blood type percentages for the entire world
total_percentage = df[['O+', 'A+', 'B+', 'AB+', 'O-', 'A-', 'B-', 'AB-']].mean()

# Creating the Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(total_percentage, labels=None, autopct='%1.1f%%')
plt.title('Blood Type Distribution Worldwide', fontsize=16)

# Create the legend separately
blood_types = ['O+', 'A+', 'B+', 'AB+', 'O-', 'A-', 'B-', 'AB-']
legend_labels = [f'{blood_type}: {percentage:.1f}%' for blood_type, percentage in zip(blood_types, total_percentage)]
plt.legend(legend_labels, loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)

plt.axis('equal')
plt.tight_layout()
plt.show()
```
