```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_data.csv')
```

```python
df.sample(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x=df['Year of Admission'])
plt.title('Students Based on Year of Admission')
```

```python
df['Year of Admission'].value_counts()
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x = df['Field of Study'],hue=df['Year of Admission'])
plt.title("Branch preference year wise")
```

```python
plt.figure(figsize = (12,4))
sns.countplot(x = df['Specialization'],hue=df['Year of Admission'])
plt.title("Specialization preference year wise")
```

```python
plt.plot(df['Fees'], df['Discount on Fees'])
plt.xlabel('Actual Fees')
plt.ylabel('Discount fees')
plt.title('Discounted fees')
```

```python
plt.scatter(df['Fees'], df['Field of Study'])
plt.xlabel('Actual Fees')
plt.ylabel('Field of Study')
plt.title('Field of Study vs Actual Fees')
```

```python
sns.pairplot(df,height=2)
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x=df['Expected Year of Graduation'])
plt.title('Students Based on Expected Year of Graduation')
```

```python
df['Expected Year of Graduation'].value_counts()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x=df['Date of Birth'])
plt.title('Students Based on Year of Birth')
plt.xlabel('Year of Birth')
```

```python
clean_df = df[df['Year of Admission'] < df['Expected Year of Graduation']]
```

```python
clean_df.sample(5)
```

```python
pd.crosstab(clean_df['Year of Admission'], clean_df['Field of Study'])
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x = clean_df['Field of Study'])
plt.title('Field of Study countplot')
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x=clean_df['Year of Admission'])
plt.title('Students Based on Year of Admission')
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x = clean_df['Field of Study'],hue=clean_df['Year of Admission'])
plt.title("Branch preference year wise")
```

```python
plt.figure(figsize = (12,4))
sns.countplot(x = clean_df['Specialization'],hue=clean_df['Year of Admission'])
plt.title("Specialization preference year wise")
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(x=clean_df['Fees'], y=clean_df['Discount on Fees'], hue=clean_df['Field of Study'])
plt.title('Fees vs Discount on Fees')
```

```python
plt.figure(figsize = (20,6))
plt.subplot(1,2,1)
sns.lineplot(x = 'Year of Admission',y='Fees',data=clean_df, ci=None)
plt.title('Year of Admission vs Fees')
plt.subplot(1,2,2)
sns.lineplot(x = 'Expected Year of Graduation',y='Fees',data=clean_df)
plt.title('Expected Year of Graduation vs Fees')
```
